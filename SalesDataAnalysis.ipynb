```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfs=pd.read_csv('sales_data.csv',encoding='ISO-8859-1')
dfs
```

```python
dfs.info()
```

```python
dfs[dfs['Amount'].isnull()]
```

```python
dfs=dfs.drop(['Status','unnamed1'],axis=1)
dfs
```

```python
dfs.dropna(inplace=True)
dfs
```

```python
d={'Marital_Status':'Marital'}
dfs.rename(columns=d,inplace=True)
dfs
```

```python
dfs['Marital']=dfs['Marital'].astype(str)
```

```python
dfs['Marital']=dfs['Marital'].str.replace('0','single')
dfs['Marital']=dfs['Marital'].str.replace('1','married')
dfs
```

```python
dfs['Gender']=dfs['Gender'].str.replace('F','Female')
dfs['Gender']=dfs['Gender'].str.replace('M','Male')
dfs
```

```python
dfs.info()
```

```python
plt.figure(figsize=(20,20),facecolor='Cyan')
plt.gcf().suptitle('Sales Data',fontsize=40,color='red',fontstyle='italic')

plt.subplot(2,3,1)
plt.title('Count of Gender')
a=sns.countplot(x='Gender',data=dfs,color='red')
plt.gca().set_facecolor('green')
for i in a.containers:
    a.bar_label(i)

plt.subplot(2,3,2)
plt.title('Perecentage of Gender')
gc=dfs['Gender'].value_counts()
plt.pie(gc,labels=gc.index,autopct='%0.2f%%',colors=['red','pink'],shadow=True,explode=[0.1,0.1])

plt.subplot(2,3,3)
gs=dfs.groupby('Gender',as_index=False)['Amount'].sum()
a=sns.barplot(x='Gender',y='Amount',data=gs,edgecolor='black',hue='Gender',palette='Set1')
for i in a.containers:
    a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('pink')
plt.xticks(rotation=45)
plt.title('Total Spending Amount by Gender')
plt.ylabel('Total Amount Spending')

plt.subplot(2,3,4)
gs=dfs.groupby('State',as_index=False)['Amount'].sum()
a=sns.barplot(y='State',x='Amount',data=gs,edgecolor='black',hue='State',palette='coolwarm')
for i in a.containers:
    a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('orange')
plt.xticks(rotation=45)
plt.title('Total Spending Amount by State')
plt.ylabel('Total Amount Spent')

plt.subplot(2,3,5)
gs=dfs.groupby('Age Group',as_index=False)['Amount'].sum()
a=sns.barplot(x='Age Group',y='Amount',data=gs,edgecolor='black',hue='Age Group',palette='rainbow')
for i in a.containers:
    a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('teal')
plt.xticks(rotation=45)
plt.title('Total Spending Amount by Different Age group  ')
plt.ylabel('Total Amount Spent')

plt.subplot(2,3,6)
gs=dfs.groupby('Occupation',as_index=False)['Amount'].sum()
a=sns.barplot(y='Occupation',x='Amount',data=gs,edgecolor='black',hue='Occupation',palette='coolwarm')
for i in a.containers:
    a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('violet')
plt.xticks(rotation=45)
plt.title('Total Spending Amount by Different Occupation')
plt.ylabel('Total Amount Spent')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,10),facecolor='Cyan')
plt.gcf().suptitle('Sales Data',fontsize=40,color='red',fontstyle='italic')


plt.subplot(1,3,1)
gs=dfs.groupby('Product_Category',as_index=False)['Amount'].sum()
a=sns.barplot(y='Product_Category',x='Amount',data=gs,edgecolor='black',hue='Product_Category',palette='Set1')
for i in a.containers:
    a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('yellow')
plt.xticks(rotation=45)
plt.title('Total Spending Amount by Different Product_Category')
plt.ylabel('Total Amount Spent')

plt.subplot(1,3,2)
gs=dfs.groupby('Marital',as_index=False)['Amount'].sum()
a=sns.barplot(x='Marital',y='Amount',data=gs,edgecolor='black',hue='Marital',palette='Set2')
for i in a.containers:
    a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('purple')
plt.xticks(rotation=45)
plt.title('Total Spending Amount by Different Marital	People')
plt.ylabel('Total Amount Spent')

plt.subplot(1,3,3)
plt.title('Age Percentage')
gc=dfs['Age Group'].value_counts()
plt.pie(gc,labels=gc.index,autopct='%0.2f%%',colors=['red','pink','orange','yellow','green','lightgreen','grey'],explode=[0.1,0,0,0,0,0,0],shadow=True)

plt.tight_layout()
plt.show()
```
